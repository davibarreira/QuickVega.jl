```julia
using Revise, Pkg
Pkg.activate("../.")
using QuickVega, DataFrames, VegaLite, VegaDatasets
using Distributions, KernelDensity
using Contour
using Setfield
```

```julia
μ = [1.0,2.0]
Σ = [3.0 2.0;
     2.0 2.0]
d = Distributions.MultivariateNormal(μ,Σ)
samples = rand(d,200)';
df = DataFrame(x =rand(10),y=rand(10))
s = QuickVega.scatterplot(samples,color="red")
```

```julia
s = QuickVega.scatterplot(df,:x,:y,color="red")
```

```julia
p = contourplot(samples[:,1],samples[:,2],color_scheme="greens")
```
